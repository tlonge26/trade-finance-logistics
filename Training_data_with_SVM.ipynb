```python
import pandas as pd
#Reading of the data cell
data = pd.read_excel("/content/Week6 - WellsFargo_Unstructured (1).xlsx")
data
```

```python
"""Using regular expression in python allows us to remove unwanted characters.
I imported the re from python which is the regular expression. I created a funtion
called CleanText which takes in the dataset that I want to clean. The first line
that I checked if the data is a test and a string value, if it is then I used a
regualr expression to remove HTTP,spaces, special character.   """

import re
def clean_text(text):
    if not isinstance(text, str):
        text = str(text)

    # Remove Hyperlinks
    text = re.sub(r'http\S+', '', text)

    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)

    # Remove Foreign Language Characters
    text = re.sub(r'[^\x00-\x7F]+', '', text)

    # Remove Spaces and Whitespaces
    text = ' '.join(text.split())

    return text
"""I created a column in my dataset called CleanedData and I applied the clenText
function that I created above on the messageText column which allows my data to
be cleaned. I saved my CleanedData set with a name called clean_data.csv"""
data['CleanedData'] = data['messageText'].apply(clean_text)
#data.to_csv('cleaned_data.csv', index=False)
```

```python
import pandas as pd
data = pd.read_excel('/content/Week6_fraud_data_small.xlsx')
data

```

```python
data.columns
```

```python
data = data.drop(['trans_date_trans_time','dob','trans_num','unix_time'],axis = 1)
data.columns
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix




# Data Cleaning and Feature Engineering
# Convert categorical variables to numeric using LabelEncoder
label_encoder = LabelEncoder()

# Assuming 'merchant', 'category', 'gender', and 'job' are categorical variables
categorical_columns = ['merchant', 'category', 'gender', 'job']

for col in categorical_columns:
    data[col] = label_encoder.fit_transform(data[col])

# Drop unnecessary columns (e.g., 'first', 'last', 'street', etc.)
data = data.drop(['first', 'last', 'street', 'city', 'state'], axis=1)

data = data.fillna(0)

# Split the data into features (X) and target variable (y)
X = data.drop('is_fraud', axis=1)
y = data['is_fraud']

# Check unique values in the target variable
unique_values = y.unique()
print(f'Unique values in the target variable: {unique_values}')

# Handle imbalanced classes (if necessary)
# You can use techniques like oversampling, undersampling, or class weights
# Example: Using class weights
class_weights = dict(zip(unique_values, len(y) / (len(unique_values) * pd.Series(y).value_counts())))
print(f'Class Weights: {class_weights}')

from sklearn.datasets import fetch_openml
import numpy as np
mnist = fetch_openml('mnist_784', version=1, as_frame=False)
X, y = mnist["data"], mnist["target"]
y = y.astype(np.uint8)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

# Standardize the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Train the SVM model
svm_model = SVC(kernel='linear', C=1.0, class_weight=class_weights, random_state=42)

svm_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = svm_model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

print(f'Accuracy: {accuracy}')
print(f'Confusion Matrix:\n{conf_matrix}')
print(f'Classification Report:\n{classification_rep}')
```
